```python
#Question 7

df_lotus=df[
    (df.make=='lotus')
]
df_lotus.head()

df_lotus_2=df_lotus[['engine_hp','engine_cylinders']]
df_lotus_2=df_lotus_2.drop_duplicates()
df_lotus_2.head()

x=np.array(df_lotus_2)
x

xt=np.transpose(x)
xt
xtx=xt.dot(x)
xtx

y=np.array([1100, 800, 750, 850, 1300, 1000, 1000, 1300, 800])
y

#Multiply the inverse of XTX with the transpose of X, and then multiply the result by y. Call the result w

xtx_inv = np.linalg.inv(xtx)
xtx_inv
h=x.dot(xtx_inv)
h
w=y.dot(h)
w
```
